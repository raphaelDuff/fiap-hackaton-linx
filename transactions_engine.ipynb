```python
import os
import pandas as pd

files = os.listdir('./transacoes')
df = pd.DataFrame()
for filename in files:
    if filename.endswith(".csv.gz"):
        df = pd.concat([pd.read_csv(
            './transacoes/'+filename, 
            compression='gzip',
            header=0,
            error_bad_lines=False,
            usecols=['encrypted_domain', 'encrypted_cnpj','encrypted_saleid','dateTime','productTotal','prod_fullname','encrypted_buyer_cpf','encrypted_buyer_cnpj']
        ), df])
        
        
# df.dropna(inplace=True) # Remove muitos dados
# df.head()
df
```
